```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
corpus = []
for i in range(len(data)):
    C# split into tokens 
    lm = WordNetLemmatizer() # lemmatizer object creation
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not') # not stopword is removed as it deprecates an emotion or negative feeling.
    review = [lm.lemmatize(word) for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
   
```

```python
# Mutliclass -
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(data['Category'])
data['Category'] = le.transform(data['Category'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()  # X mein sentences upload kiye hai
y = data['Category'] # y mein output labels mein upload kiye hai 
# X[0].shape # Vector size = 179 unique words
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.20,random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

```python
# Now saving classifier model and vectors model also we need to save it.
import pickle
pickle.dump(classifier,open('First_NLP_Model.pkl','wb'))
```

```python
pickle.dump(cv,open('vector.pkl','wb')) # Saving the vector in pickle.
```
